```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from plotnine import *

# cargamos los datos como un Datafame
data_ej3 = pd.read_csv("/content/drive/MyDrive/EUDE/Visualizacion de Datos/Base de Datos/Ejercicio 3.csv")
```

```python
# veamos la estructura del fichero
data_ej3.head()
```

```python
#visualizacion de resultados
ggplot(data_ej3, aes(x=data_ej3.Weight, y=data_ej3.Cholesterol,color=data_ej3.Gender)) + geom_point()
```

```python
#visualización y K Medias
k = 4
kmeans = KMeans(n_clusters=k)
y_pred = kmeans.fit_predict(data_ej3)
```

```python
ggplot(data_ej3, aes(x=data_ej3.Weight, y=data_ej3.Cholesterol,color='y_pred')) + geom_point()
```

```python
kmeans.cluster_centers_
```

```python
#vemos con exactitud los centroides:
kmeans.cluster_centers_
z=kmeans.cluster_centers_[:, :2]
plt.scatter(z[:, 0], z[:, 1])
```

```python
ggplot(data_ej3, aes(x="Weight", y="Cholesterol", color="y_pred")) + geom_point()+ facet_wrap("Gender")
```

```python
graf1 = ggplot(data_ej3, aes(x="Weight", y="Cholesterol",color="y_pred")) + geom_point() + facet_wrap("Gender")
graf1
```
